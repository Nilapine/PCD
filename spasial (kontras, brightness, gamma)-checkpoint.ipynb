# RGB TO Greyscale

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
image = cv2.imread('bahan/lenna.jpg')  # Ganti 'your_image.jpg' dengan nama gambar Anda

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)


# Reduce to 1-bit grayscale (black and white)
gray_1bit = np.where(gray > 127, 255, 0).astype(np.uint8)  # Threshold at 127
print(image.shape)
print(gray.shape)
print(gray_1bit.shape)


# Convert to grayscale

# Plot RGB to Grayscale
plt.figure(figsize=(8, 4))
plt.subplot(131)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original (RGB)')
plt.subplot(132)
plt.imshow(gray, cmap='gray')
plt.title('Grayscale 8 bit')
plt.subplot(133)
plt.imshow(gray_1bit, cmap='gray')
plt.title('Grayscale 1 bit')
plt.show()

```

# Invert

```python

# Invert grayscale image
inverted = 255 - gray

# Plot Inverted Image
plt.figure(figsize=(8, 4))
plt.subplot(121)
plt.imshow(gray, cmap='gray')
plt.title('Original Grayscale')
plt.subplot(122)
plt.imshow(inverted, cmap='gray')
plt.title('Inverted')
plt.show()
```

# Brightness

```python

# 1. Brightness Adjustment

# Brightness +50
bright_plus = cv2.add(gray, 50) 

# Plot Brightness +50
plt.figure(figsize=(8, 4))
plt.subplot(121)
plt.imshow(gray, cmap='gray')
plt.title('Original')
plt.subplot(122)
plt.imshow(bright_plus, cmap='gray')
plt.title('Brightness +50')
plt.show()

# Brightness -50
bright_minus = cv2.subtract(gray, 50)

# Plot Brightness -50
plt.figure(figsize=(8, 4))
plt.subplot(121)
plt.imshow(gray, cmap='gray')
plt.title('Original')
plt.subplot(122)
plt.imshow(bright_minus, cmap='gray')
plt.title('Brightness -50')
plt.show()
```

# Contrast

```python
# 2. Contrast Adjustment
# Contrast +10 (nilai grayscale pusat = 50)
contrast_plus = cv2.addWeighted(gray, 1.2, np.zeros(gray.shape, gray.dtype), 0, 50)  # 50 = (1.2 - 1) * 50


# Plot Contrast +2
plt.figure(figsize=(8, 4))
plt.subplot(121)
plt.imshow(gray, cmap='gray')
plt.title('Original')
plt.subplot(122)
plt.imshow(contrast_plus, cmap='gray')
plt.title('Contrast +10')
plt.show()

# Contrast -2
contrast_minus = cv2.addWeighted(gray, 0.8, np.zeros(gray.shape, gray.dtype), 0, -50)  # -50 = (0.8 - 1) * 50

# Plot Contrast -2
plt.figure(figsize=(8, 4))
plt.subplot(121)
plt.imshow(gray, cmap='gray')
plt.title('Original')
plt.subplot(122)
plt.imshow(contrast_minus, cmap='gray')
plt.title('Contrast -10')
plt.show()
```

```python
# 3. Gamma Correction


gamma_2 = np.power(gray / 255.0, 2) * 255.0
gamma_2 = gamma_2.astype(np.uint8)


# Plot Gamma 1
plt.figure(figsize=(8, 4))
plt.subplot(121)
plt.imshow(gray, cmap='gray')
plt.title('Original')
plt.subplot(122)
plt.imshow(gamma_2, cmap='gray')
plt.title('Gamma 2')
plt.show()

# Gamma 0.5
gamma_05 = np.power(gray / 255.0, 0.5) * 255.0
gamma_05 = gamma_05.astype(np.uint8)

# Plot Gamma 0.5
plt.figure(figsize=(8, 4))
plt.subplot(121)
plt.imshow(gray, cmap='gray')
plt.title('Original')
plt.subplot(122)
plt.imshow(gamma_05, cmap='gray')
plt.title('Gamma 0.5')
plt.show()
```

# Histogram citra

```python
# Calculate histogram of original grayscale image
hist_gray = cv2.calcHist([gray], [0], None, [256], [0, 256])

# Calculate histogram of gamma-corrected image
hist_kontras1 = cv2.calcHist([contrast_plus], [0], None, [256], [0, 256])


# Calculate histogram of gamma-corrected image
hist_kontras2 = cv2.calcHist([contrast_minus], [0], None, [256], [0, 256])

# Plot histograms
plt.figure(figsize=(18, 4))
plt.subplot(131)
plt.plot(hist_gray)
plt.title('Histogram (Original)')
plt.xlabel('Intensity')
plt.ylabel('Frequency')
plt.subplot(132)
plt.plot(hist_kontras1)
plt.title('Histogram Contrast +10')
plt.xlabel('Intensity')
plt.ylabel('Frequency')
plt.subplot(133)
plt.plot(hist_kontras2)
plt.title('Histogram Contrast -10')
plt.xlabel('Intensity')
plt.ylabel('Frequency')
plt.show()
```
